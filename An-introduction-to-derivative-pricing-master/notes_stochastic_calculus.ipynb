## **<center><font color=blue>伊藤积分</font></center>**
简单过程$\Delta(t)$的伊藤积分，记为：

$$I(t)=\int_{0}^{t}\Delta (u)dW(u)$$

连续变化的被积函数$\Delta(t)$的伊藤积分定义为:

$$\int_{0}^{t}\Delta(u)dW(u)=\lim_{n \to \infty}\Delta_n(u)dW(u)$$

以上定义的$I(t)=\int_{0}^{t}\Delta(u)dW(u)$积分具有以下性质:

1. **(连续性)**作为积分上限t的函数，I(t)的路径连续.
2. **(适应性)**对每个他，I(t)为$\mathcal{F}(t)$-可测
3. **(线性性)**
4. **(鞅性质)**
5. **(伊藤等距)** $EI^2(t)=E\int_{0}^{t}\Delta^2(u)du$
6. **(二次变差)** $[I,I](t)=\int_{0}^{t}\Delta^2(u)du$

## **<center>伊藤过程 (or Ito's process or stochastic process)</center>**  
$$X_t=X_0+\int_{0}^{t}\sigma_sdW_s+\int_{0}^{t}\mu_sds$$
the differential form can be written:  
$$dX_t=\sigma_tdW_t+\mu_tdt$$

<font color=red>注意：先定义的伊藤积分，再据此有微分形式</font>

In the special case when $\sigma$ and $\mu$ depend on W onlu through $X_t$,such as 
$\sigma_t=\sigma(X_t,t)$,where $\sigma(x,t)$ is some deterministic function,the equation 
$$dX_t=\sigma(X_t,t)dW_t+\mu(X_t,t)dt$$
is called a stochastic differential equation (SED) for X.


## **<center>Ito's formula(伊藤-德布林公式)  伊藤引理</center>**
对于布朗运动的函数$f(W_t)$求导数或者说求微分的时候，公式与普通微积分公式不一样，这个求微分的公式，就是伊藤引理(Ito lemma)

If $Y_t:=f(t,X_t)$,then it's SDE is:
$$dY_t=f_t(t,X_t)d_t+f_X(t,X_t)dX_t+\frac{1}{2}f''(t,X_t)(dX_t)^2$$

# <center><font color=blue>quantos</font></center>

A [quanto](https://en.wikipedia.org/wiki/Quanto) is a type of derivative in which the underlying is denominated in one currency, but the instrument itself is settled in another currency at some rate.For example:
* a digital contract which pays one dollar at time T if the then British Petroleum stock price is larger than some pre-agreed stike;

* a forward contract,namely receiving the BP stock price at time T as if it were in dollars in exhange for paying a pre-agreed dollar amount;

* an option to receive the BP stock price less a strike price,in dollars.

### quant model  
>sterling stock price $S_t=S_0e^{\sigma_1W_1(t)+\mu t}$  
the value of one pound in dollars $C_t=C_0e^{\bar {\rho}\sigma_2W_t(t)+vt}$  
dollar cash bond $B_t=e^{rt}$  
sterling cash bond $D_t=e^{ut}$


In this model,here are three tradables:
* the dollar worth of the sterling bond:<font color=blue>$C_tD_t$</font>
* the dollar worth of the stock:<font color=blue>$C_tS_t$</font>
* the dollar cash bond:<font color=blue>$B_t$</font>

Writing down the first two of these tradables after discounting by the third,the numeraire,we have $Y_t=B_t^{-1}C_tD_t$ and $Z_t=B_t^{-1}C_tS_T$ respectively.Their SDEs are:
$$dY_t=Y_t(\rho \sigma_2dW_1(t) + \bar {\rho} \sigma_2 dW_2(t) + (v+\frac{1}{2}\sigma_2^2 +\mu -r)dt)$$
$$dZ_t=Z_t((\sigma_1+\rho \sigma_2)dW_1(t)+\bar {\rho} \sigma_2 dW_2(t) +(\mu +v +\frac{1}{2} +\rho \sigma_1\sigma_2+\frac{1}{2}\sigma_2^2-r)dt)$$

where $\rho\in [-1,1]$ and $\bar \rho=\sqrt{1-\rho^2}$  
(This can be checked using the n-factor Ito's formula of section 6.3)

3. Find a previsible process $\phi_t$,such that $dE_t=\phi_tdS_t$

### pricing  
**forward contract**  
The forward at time zero in dollars is  
$$
\begin{align}
V_0 & =e^{-rT}E_Q(S_T-k)\\
    & =e^{-rT}E_Q(S_0e^{\sigma_1\sqrt{T}Z +(\mu -\rho \sigma_1 \sigma_2 -\frac{1}{2}\sigma_1^2)T}-k)\\
    & =e^{-rT}(S_0e^{(\mu +\rho \sigma_1 \sigma_2)T}-k)
\end{align}
$$  
where $Z\sim N(0,1)$ under Q.  


For this to be on market,that is to have a value of zero,we must set k to be $S_0e^{(\mu +\rho \sigma_1 \sigma_2)T}$.

>Implication:
We know that for the local currency forward $F=S_0e^{\mu T}$,so $k=\color{blue}{F}e^{-\rho \sigma_1 \sigma_2 T}$.It is clear that this quanto forward price is greater than the simple forward price if and only if the correlation between the stock and the exchange rate is negative.
