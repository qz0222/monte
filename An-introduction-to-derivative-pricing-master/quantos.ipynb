## Background 

A [quanto](https://en.wikipedia.org/wiki/Quanto) is a type of derivative in which the underlying is denominated in one currency, but the instrument itself is settled in another currency at some rate.For example:
* a digital contract which pays one dollar at time T if the then British Petroleum stock price is larger than some pre-agreed stike;

* a forward contract,namely receiving the BP stock price at time T as if it were in dollars in exhange for paying a pre-agreed dollar amount;

* an option to receive the BP stock price less a strike price,in dollars.  


## Assumption  
> 
sterling stock price $S_t=S_0e^{\sigma_1W_1(t)+\mu t}\tag{1}$  
the value of one pound in dollars $C_t=C_0e^{\bar {\rho}\sigma_2W_t(t)+vt}\tag{2}$  
dollar cash bond $B_t=e^{rt} \tag{3}$  
sterling cash bond $D_t=e^{ut} \tag{4}$
*where $\rho\in [-1,1]$ and $\bar \rho=\sqrt{1-\rho^2}$* 


## Change of Measure
In this model,here are three tradables:  
* the dollar worth of the sterling bond:***$C_tD_t$***
* the dollar worth of the stock:***$C_tS_t$***
* the dollar cash bond:***$B_t$***

Writing down the first two of these tradables after discounting by the third,the numeraire,we have $Y_t=B_t^{-1}C_tD_t$ and $Z_t=B_t^{-1}C_tS_T$ respectively.Their SDEs are:
$$dY_t=Y_t(\rho \sigma_2dW_1(t) + \bar {\rho} \sigma_2 dW_2(t) + (v+\frac{1}{2}\sigma_2^2 +\mu -r)dt) \tag{5}
$$
$$
dZ_t=Z_t((\sigma_1+\rho \sigma_2)dW_1(t)+\bar {\rho} \sigma_2 dW_2(t) +(\mu +v +\frac{1}{2} +\rho \sigma_1\sigma_2+\frac{1}{2}\sigma_2^2-r)dt) \tag{6}
$$
<br/>
where $\rho\in [-1,1]$ and $\bar \rho=\sqrt{1-\rho^2}$  
(This can be checked using the n-factor Ito's formula of section 6.3)  
[multiple stock models](multiple_stock_models.ipynb)

According to n-factor GCM,we need to find $\gamma_t=(\gamma_1(t),\gamma_2(t))$ to make:
$$dY_t=Y_t\left(\rho \sigma_2 d \tilde{W}_1(t) + \bar {\rho} \sigma_2 d \tilde {W}_2(t)\right) \tag {7}$$
$$dZ_t=Z_t\left((\sigma_1+\rho \sigma_2) d \tilde{W}_1(t)+\bar {\rho} \sigma_2  d \tilde{W}_2(t)\right) \tag{8}$$  
Where $ d\tilde {W}_1(t)=dW_1(t)+\rho_1(t)dt$ and $d\tilde {W}_2(t)=dW_2(t)+\rho_2(t)dt$

Thus under Q we can write the original process $S_t$ and $C_t$ as:
$$
S_t=S_0e^{\sigma_1\tilde W_1(t) +(\mu -\rho \sigma_1 \sigma_2 -\frac{1}{2}\sigma_1^2)t} \tag{9}
$$ 
$$C_t=C_0e^{\rho \sigma_2 \tilde W_2(t) +(r-\mu-\frac{1}{2}\sigma_2^2)t)} \tag{10}
$$


## Pricing
**forward contract**  
The forward at time zero in dollars is  
$$
\begin{align}
V_0 & =e^{-rT}E_Q(S_T-k)\\
    & =e^{-rT}E_Q(S_0e^{\sigma_1\sqrt{T}Z +(\mu -\rho \sigma_1 \sigma_2 -\frac{1}{2}\sigma_1^2)T}-k)\\
    & =e^{-rT}(S_0e^{(\mu +\rho \sigma_1 \sigma_2)T}-k)
\end{align} \tag{11}
$$  
where $Z\sim N(0,1)$ under Q.  


For this to be on market,that is to have a value of zero,we must set k to be $S_0e^{(\mu {\color {red}{+\rho \sigma_1 \sigma_2}})T}$.


## **Implication**
We know that for the local currency forward $F=S_0e^{\mu T}$,so $k=\color{blue}{F}e^{-\rho \sigma_1 \sigma_2 T}$.It is clear that this quanto forward price is greater than the simple forward price if and only if the correlation between the stock and the exchange rate is negative.
